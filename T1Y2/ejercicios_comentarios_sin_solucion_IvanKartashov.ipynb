# Ejercicios sobre Comentarios en Python
Este notebook contiene **5 ejercicios** para practicar cómo escribir comentarios en Python:

- Comentarios al final de línea
- Comentarios en varias líneas

- Desactivar código comentándolo.

## 1️⃣ Añadir comentarios descriptivos
**Enunciado:**
El siguiente código no tiene ninguna explicación. Añade **comentarios con `#`** para que quede claro qué hace cada línea.
```python
x = 8
y = 3
print(x * y)
```

````python
#This line will define the variable x and assign the value of 8 to it
x = 8 
#This line defines the variable y with the initial value of 3
y = 3 
#print is used to display lines of text, in this case print will display the result of multiplication of the variable x and the variable y
print(x * y) 
````

## 2️⃣ Comentario al final de la línea
**Enunciado:**
Agrega un comentario al final de cada instrucción explicando su función.
```python
a = 5
b = 2
print(a / b)
```

````python
a = 5 #This line defines variable a and assigns it the value of 5
b = 2 #This line defines variable b and assigns it the value of 2
print (a / b) #This line will print the outcome of divison beetween a and b, but since the variables are defined as integer it will round up the answer
````

## 3️⃣ Explicación en varias líneas
**Enunciado:**
Escribe un **comentario de varias líneas con `#`** al inicio que explique lo que hace este programa.
```python
nombre = input("Introduce tu nombre: ")
print("Hola", nombre)
```

````python
#We define the variable nombre and se it to input, letting the user choose its value in the terminal
nombre = input("Introduce tu nombre: ")
#And the next line we shall use to print the outcome on-screen with the name of the user and a greeting 
print("Hola", nombre)
````

## 4️⃣ Docstring en función
**Enunciado:**
Completa el **docstring** de esta función para explicar qué hace y qué devuelve.
```python
def cuadrado(n):

```

````python
def cuadrado(n):

    """Creates a function of cuadrado which returns the value of n **2"""

    return n ** 2

print(cuadrado(5))

help(cuadrado)
````
