### 1Ô∏è‚É£ Suma de los n√∫meros del 1 al 50 que sean m√∫ltiplos de 5
**Problema:** Escribe una funci√≥n `suma_multiplos_5()` que calcule la suma de todos los m√∫ltiplos de 5 entre 1 y 50 (ambos incluidos).

```python
def suma_multiplos_5():
    sumamultiplos = 0
    for i in range(1, 51):
        if i % 5 == 0:
            sumamultiplos += i
    return sumamultiplos

```

```python
assert suma_multiplos_5() == 275
```

**Problema:** Crea una funci√≥n `cuenta_regresiva()` que devuelva una cadena que muestre los n√∫meros del 10 al 0 (en orden descendente).

### 3Ô∏è‚É£ Contar cu√°ntos n√∫meros impares hay entre dos valores dados
**Problema:** Implementa una funci√≥n `contar_impares(inicio, fin)` que cuente cu√°ntos n√∫meros impares hay entre los valores dados (inclusive).

```python
def contar_impares(inicio, fin):
    nimpares = 0
    inicioreal = inicio - 1
    finreal = fin + 1
    for i in range (inicioreal, finreal):
        if i % 2 == 1:
            nimpares += 1
    return nimpares
```

```python
assert contar_impares(1,10) == 5
assert contar_impares(3,7) == 3
```

```python
def sin_multiplos_de_4(x):
    for i in range(1, x+1, 1):
        cadena = ""
        if i % 4 == 0:
            continue
        else:
            cadena += str(i) + " "
    return cadena

print(sin_multiplos_de_4(21))
```

### 6Ô∏è‚É£ Clasificar nota con `match`
**Problema:** Usa `match` para devolver la calificaci√≥n textual de una nota (0‚Äì10):
- 0‚Äì4 ‚Üí 'Insuficiente'
- 5 ‚Üí 'Suficiente'
- 6 ‚Üí 'Bien'
- 7‚Äì8 ‚Üí 'Notable'
- 9‚Äì10 ‚Üí 'Sobresaliente'

```python
def calificacion(nota):
    match nota:
        case (nota) if nota <= 4:
            return "Insuficiente"
        case (nota) if nota == 5:
            return "Suficiente"
        case (nota) if nota == 6:
            return "Bien"
        case (nota) if nota == 7 or nota == 8:
            return "Notable"
        case (nota) if nota == 10:
            return "Sobresaliente"
```

```python
assert calificacion(3) == 'Insuficiente'
assert calificacion(8) == 'Notable'
assert calificacion(10) == 'Sobresaliente'
```

### 8Ô∏è‚É£ Contar cu√°ntos n√∫meros negativos hay entre dos n√∫meros
**Problema:** Crea una funci√≥n `contar_negativos(lista)` que devuelva cu√°ntos n√∫meros son negativos.

```python
def contar_negativos(inicio, fin):
    contadornegativos = 0
    for i in range(inicio, fin):
        if i < 0:
            contadornegativos += 1
    return contadornegativos
```

```python
assert contar_negativos(-6, 8) == 6
assert contar_negativos(-3, 8) == 3
assert contar_negativos(6, 8) == 0
```

- Factorial 4: `4*3*2*1`
- Factorial 3: `3*2*1`

```python
n = 5
factorialcero = 0
factorial = 0
if n != 0:
    for i in range(0, n+1):
        factoriali = n * (n - i)
        factorial += factoriali
    print (factorial)
else:
    print (factorialcero)
    
```

```python
def factorial(n):
    factorialcero = 0
    factorial = 0
    if n != 0:
        for i in range(n):
            factoriali = n * (n - i)
            factorial += factoriali
    else:
        return factorialcero
    return factorial
    
```

```python
assert factorial(5) == 120
assert factorial(6) == 720
assert factorial(4) == 24
assert factorial(3) == 6
assert factorial(0) == 1
```
