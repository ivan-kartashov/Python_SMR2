# Tema 4. Tipos avanzados de datos. Cadenas y Listas en Python

Esta hoja incluye ejercicios para practicar toda la teoría vista sobre cadenas de caracteres y listas en Python. Cada ejercicio tiene **tests unitarios** que permiten comprobar si la solución es correcta.

```python
## EJERCICIO 1: Contar vocales

def contar_vocales(texto):
    voc = ["a", "e", "i", "o", "u", "A", "E", "I", "O", "U"]
    contador = 0
    for letra in texto:
        if letra in voc:
            contador += 1
        else:
            continue
    return contador

# Tests
assert contar_vocales('hola') == 2
assert contar_vocales('Python') == 1
assert contar_vocales('AEIOU') == 5
assert contar_vocales('') == 0
```

```python
## EJERCICIO 2: Invertir cadena

def invertir_cadena(texto):
    texto_inverso = ""
    contador = -1
    for letra in texto:
        texto_inverso += texto[contador]
        contador -= 1
    return texto_inverso

# Tests
assert invertir_cadena('hola') == 'aloh'
assert invertir_cadena('Python') == 'nohtyP'
assert invertir_cadena('') == ''
```

```python
## EJERCICIO 3: Subcadena

def contiene_subcadena(cadena, subcadena):
    if subcadena in cadena:
        return True
    else:
        return False

# Tests
assert contiene_subcadena('hola mundo', 'mundo') == True
assert contiene_subcadena('Python', 'thon') == True
assert contiene_subcadena('Python', 'java') == False
assert contiene_subcadena('', '') == True
```

```python
## EJERCICIO 4: Convertir a mayúsculas

def a_mayusculas(texto):
    texto_mayus = texto.upper()
    return texto_mayus

# Tests
assert a_mayusculas('python') == 'PYTHON'
assert a_mayusculas('Hola Mundo') == 'HOLA MUNDO'
assert a_mayusculas('') == ''
```

```python
## EJERCICIO 5: Convertir a minúsculas

def a_minusculas(texto):
    texto_minux = texto.lower()
    return texto_minux

# Tests
assert a_minusculas('PYTHON') == 'python'
assert a_minusculas('Hola Mundo') == 'hola mundo'
assert a_minusculas('') == ''
```

```python
## EJERCICIO 6: Reemplazar subcadena

def reemplazar_subcadena(texto, antiguo, nuevo):
    if antiguo in texto:
        return texto.replace(antiguo, nuevo)
    else:
        return ""

# Tests
assert reemplazar_subcadena('hola mundo', 'mundo', 'Python') == 'hola Python'
assert reemplazar_subcadena('aaaa', 'a', 'b') == 'bbbb'
assert reemplazar_subcadena('', 'a', 'b') == ''
```

```python
## EJERCICIO 7: Eliminar espacios al principio y final

def limpiar_espacios(texto):
    return texto.strip()

# Tests
assert limpiar_espacios('  hola  ') == 'hola'
assert limpiar_espacios('  Python  ') == 'Python'
assert limpiar_espacios('') == ''
```

```python
## EJERCICIO 8: Dividir cadena

def dividir_cadena(texto, separador):
    return texto.split(separador)

# Tests
assert dividir_cadena('uno dos tres', ' ') == ['uno','dos','tres']
assert dividir_cadena('manzana,pera,banana', ',') == ['manzana','pera','banana']
assert dividir_cadena('', ',') == ['']
```

```python
## EJERCICIO 9: Concatenar lista de cadenas

def concatenar_cadenas(lista_cadenas):
    return " ".join(lista_cadenas)

# Tests
assert concatenar_cadenas(['hola','mundo']) == 'hola mundo'
assert concatenar_cadenas(['Python','es','genial']) == 'Python es genial'
assert concatenar_cadenas([]) == ''
```

```python
## EJERCICIO 10: Sumar elementos de una lista

def sumar_lista(lista):
    suma_total = 0
    for x in lista:
        suma_total += x
    return suma_total

# Tests
assert sumar_lista([1,2,3,4]) == 10
assert sumar_lista([]) == 0
assert sumar_lista([10,-5,3]) == 8
```

```python
## EJERCICIO 11: Filtrar elementos mayores que n

def mayores_que(lista, n):
    lista_final = []
    for valor in lista:
        if valor > n:
            lista_final.append(valor)
    return lista_final


# Tests
assert mayores_que([1,5,10,20],5) == [10,20]
assert mayores_que([0,-1,-5],-2) == [0,-1]
assert mayores_que([],10) == []
```

```python
## EJERCICIO 12: Añadir elemento al final (append)

def agregar_final(lista, elemento):
    lista.append(elemento)

# Tests
l=[1,2]
agregar_final(l,3)
assert l == [1,2,3]

l=[]
agregar_final(l,'a')
assert l == ['a']
```

```python
## EJERCICIO 13: Insertar elemento en posición

def insertar_elemento(lista, indice, elemento):
    lista.insert(indice, elemento)

# Tests
l=[1,2,4]
insertar_elemento(l,2,3)
assert l == [1,2,3,4]
```

```python
## EJERCICIO 14: Extender lista con otra lista

def extender_lista(lista, otra_lista):
    lista.extend(otra_lista)

# Tests
l=[1,2]
extender_lista(l,[3,4])
assert l == [1,2,3,4]
```

```python
## EJERCICIO 15: Eliminar elemento por valor

def eliminar_valor(lista, valor):
    lista.remove(valor)

# Tests
l=['a','b','c','b']
eliminar_valor(l,'b')
assert l == ['a','c','b']
```

```python
## EJERCICIO 16: Eliminar y devolver elemento (pop)

def eliminar_pop(lista, indice):
    return lista.pop(indice)



# Tests
l=[1,2,3]
valor=eliminar_pop(l,1)
assert valor==2
assert l==[1,3]
```

```python
## EJERCICIO 17: Vaciar lista (clear)

def vaciar_lista(lista):
    lista.clear()

# Tests
l=[1,2,3]
vaciar_lista(l)
assert l==[]
```

```python
## EJERCICIO 18: Invertir lista

def invertir_lista(lista):
    lista_inversa = []
    for elemento in lista:
        lista_inversa.insert(0, elemento)
    return lista_inversa

# Tests
assert invertir_lista([1,2,3])==[3,2,1]
assert invertir_lista([])==[]
```

```python
## EJERCICIO 19: Contar ocurrencias de un elemento

def contar_elemento(lista, elemento):
    cuenta_elemento = 0
    for element in lista:
        if element == elemento:
            cuenta_elemento += 1
        else:
            continue
    return cuenta_elemento

# Tests
assert contar_elemento([1,2,2,3,2],2)==3
assert contar_elemento([],5)==0
```

```python
## EJERCICIO 20: Sublista con slicing

def sublista(lista, start, end):
    return lista[start:end]

# Tests
assert sublista([10,20,30,40,50],1,4)==[20,30,40]
assert sublista([],0,1)==[]
```
