# Práctica de Tuplas en Python
Los `assert` permiten comprobar si la estructura es correcta.

## Ejercicios Básicos

```python
# Ejercicio 1: Crea una tupla con 3 colores favoritos
colores = ('rojo', 'azul', 'blanco')  # Pista: usa paréntesis y separa los elementos con comas
assert isinstance(colores, tuple) and len(colores)==3, "La tupla debe tener 3 elementos"
```

```python
# Ejercicio 2: Accede al primer color de la tupla
primer_color = ("rojo", "azul", "amarillo") #Para acceder al primer color hariamos print (primer_color(0))
  # Pista: usa el índice 0 para acceder al primer elemento
assert primer_color is not None, "Debes asignar el primer color"
```

```python
# Ejercicio 3: Crea una tupla con números del 1 al 5
numeros = (1, 2, 3, 4, 5)  # Pista: usa comas para separar los números, ejemplo (1,2,3,...)
assert isinstance(numeros, tuple) and len(numeros)==5, "Tupla de 5 elementos"
```

```python
# Ejercicio 4: Suma todos los elementos de la tupla de números
tupla = (1, 4, 10)
total = sum(tupla)  # Pista: puedes usar sum() sobre la tupla
assert total is not None, "Debes calcular la suma"
```

```python
# Ejercicio 5: Repite la tupla ('FP',) 4 veces
t_repetida = ("FP",)   # Pista: usa el operador * para repetir la tupla
t_repetida = t_repetida * 4
assert isinstance(t_repetida, tuple) and len(t_repetida)==4, "Tupla repetida de 4 elementos"
```

```python
# Ejercicio 6: Slicing de la tupla de números, desde el segundo hasta el cuarto elemento
sub_numeros = (1, 2, 3, 4)  # Pista: usa numeros[1:4] para obtener los elementos 2,3,4
sub_numeros = sub_numeros[1:4]
assert isinstance(sub_numeros, tuple) or isinstance(sub_numeros, list), "Debes crear un subgrupo"
```

## Ejercicios Intermedios

```python
# Ejercicio 10: Tupla anidada de coordenadas, accede a la segunda coordenada
coordenadas = ((0,0),(1,1),(2,2))
segunda_coord = (coordenadas[1])  # Pista: usa índices para acceder a la tupla anidada
assert segunda_coord is not None, "Debes asignar la segunda coordenada"
```

```python
# Ejercicio 11: Lista de tuplas alumnos y nota. Obtén nota de Marta
alumnos = [("Ana",7),("Luis",9),("Marta",8)]
for nombre, nota in alumnos:
    if nombre == "Marta":
        nota_marta = nota
        break
# Pista: recorre la lista y compara el nombre
assert nota_marta is not None, "Debes obtener la nota de Marta"
```

```python
# Ejercicio 12: Desempaquetado múltiple
t = (10,20,30,40,50)
primero, *medio, ultimo = (t) * 5  # Pista: reemplaza None con la tupla t
assert primero is not None and ultimo is not None, "Debes desempaquetar correctamente"
```

## Ejercicios Avanzados

```python
# Ejercicio 16: Filtra productos con stock>30
productos = [("Leche",1.5,20),("Pan",0.8,50),("Huevos",0.2,100)]
productos_disponibles = [
        nombre
        for nombre, precio, stock in productos if stock > 30
    ]
# Pista: usa list comprehension y condicion stock>30
assert isinstance(productos_disponibles, list), "Debes crear una lista de productos"
```

```python
# Ejercicio 17: Cambiar nota de Luis a 10
alumnos = [("Ana",7),("Luis",9),("Marta",8)]
alumnos_actualizados = []  # Pista: crea nueva lista reemplazando la nota de Luis
for nombre, nota in alumnos:
    if nombre == "Luis":
        alumnos_actualizados.append((nombre, 10))
    else:
        alumnos_actualizados.append((nombre, nota))
assert isinstance(alumnos_actualizados, list), "Debes crear lista actualizada"
```
