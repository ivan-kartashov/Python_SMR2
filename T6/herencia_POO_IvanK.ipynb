## Ejercicio 1 – Clase base Persona

Crea una clase llamada `Persona` que tenga:

```python
#Ejemplo 
class Oviparo:
    def __init__(self, nombre):
        self.nombre = nombre
        self.__huevos = True
    def poner_huevos(self):
        if(self.__huevos == True):
            print(f"-------El animal {self.nombre} pone huevos-----------")
        else:
            print(f"-------El animal {self.nombre} no pone huevos---------")
class Pez(Oviparo):
    def __init__(self, nombre, aletas):
        super().__init__(nombre)
        self.escamas = True
        self.aletas = aletas
    def nadar(self):
        print(f"------ estoy nadando ------- con {self.aletas} aletas --------------")

pez = Pez("Pez payaso", 3)
pez.poner_huevos()
pez.nadar()
```

```python
# TODO: Define la clase Persona
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad
    def mostrar_info(self):
        return f"El nombre de la persona es {self.nombre} y su edad es {self.edad} años"
# TODO: Crea un objeto Persona y prueba el método mostrar_info()
```

## Ejercicio 2 – Herencia: Alumno

Crea una clase `Alumno` que herede de `Persona` y añada:
- Un atributo `curso`
- Un método `mostrar_curso()`

```python
# TODO: Define la clase Alumno que herede de Persona
class Alumno(Persona):
    def __init__(self, nombre, edad, curso):
        super().__init__(nombre, edad)
        self.curso = curso
    def mostrar_curso(self):
        return f"Curso: {self.curso}"

# TODO: Crea un objeto Alumno y muestra toda su información
```

```python
# Tests Ejercicio 2
a = Alumno("Luis", 18, "SMR")
assert a.nombre == "Luis"
assert a.edad == 18
assert a.curso == "SMR"
assert a.mostrar_curso() == "Curso: SMR"
```

## Ejercicio 4 – Herencia y sobrescritura de métodos

Crea una clase `Profesor` que herede de `Persona` y:

- Sobrescriba el método `mostrar_info()` para que también muestre la especialidad

```python
# TODO: Define la clase Profesor
class Profesor(Persona):
    def __init__(self, nombre, edad, asignatura):
        super().__init__(nombre, edad)
        self.asignatura = asignatura
    def mostrar_info(self):
        return f"La profesora {self.nombre}, tiene {self.edad} años de edad y su asignatura es {self.asignatura}"

# TODO: Crea un Profesor y llama a mostrar_info()
```

## Ejercicio 5 – Ejercicio práctico: Vehículos

1. Crea una clase `Vehiculo` con:
   - marca
   - velocidad
   - método `acelerar()` que aumente la velocidad

2. Crea una clase `Coche` que herede de `Vehiculo` y añada:
   - número de puertas
   - método `mostrar_datos()`

```python
# TODO: Define la clase Vehiculo
class Vehiculo:
    def __init__(self, marca):
        self.marca = marca
        self.velocidad = 0
    def acelerar(self, aceleracion):
        self.velocidad += aceleracion
# TODO: Define la clase Coche que herede de Vehiculo
class Coche(Vehiculo):
    def __init__(self, marca, puertas=None):
        super().__init__(marca)
        self.puertas = puertas
    def mostrar_datos(self):
        return f"El coche es de la marca {self.marca}, su velocidad es de {self.velocidad} km/h y tiene {self.puertas} puertas"

# TODO: Crea un coche, acelera y muestra sus datos
micoche = Coche("Volkswagen", 5)
micoche.acelerar(100)
print(micoche.velocidad)
micoche.mostrar_datos()
```

```python
# Tests Ejercicio 5
c = Coche("Seat", 5)
assert c.marca == "Seat"
assert c.puertas == 5
assert c.velocidad == 0

c.acelerar(30)
assert c.velocidad == 30
assert "Seat" in c.mostrar_datos()
```
