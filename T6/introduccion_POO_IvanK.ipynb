## Ejercicio 2 â Clase Viaje

Crea la clase `Viaje` con atributos: `destino`, `dias`, `precio`. AÃąade un mÃĐtodo `coste_diario()` que devuelva el precio dividido entre los dÃ­as.

```python
# Escribe tu cÃģdigo aquÃ­
class Viaje:
    def __init__(self, destino, dias, coste):
        self.destino = destino
        self.dias = dias
        self.coste = coste
    
    def coste_diario(self):
        coste_total = self.coste / self.dias
        return coste_total
```

```python
v = Viaje("Roma", 4, 400)
assert v.coste_diario() == 100
```

## Ejercicio 4 â Clase Cancion

Crea la clase `Cancion` con atributos: `titulo`, `artista`, `duracion`. AÃąade un mÃĐtodo `es_larga()` que devuelva True si la duraciÃģn es mayor a 5 minutos.

```python
# Escribe tu cÃģdigo aquÃ­
class Cancion:
    def __init__(self, nombre, artista, duracion):
        self.nombre = nombre
        self.artista = artista
        self.duracion = duracion
    
    def es_larga(self):
        if self.duracion >= 5:
            return True
        else:
            return False
```

```python
c = Cancion("Epic Song", "DJ Max", 6)
assert c.es_larga() is True
```
