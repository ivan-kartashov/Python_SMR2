### 1. Definir funciones
Las funciones en Python son bloques de código que realizan una tarea específica y pueden reutilizarse. Ayudan a estructurar el programa y evitar duplicar código.

Se definen usando la palabra clave `def`. Una función puede no recibir parámetros y no devolver nada.

#### Ejemplo 1
Función simple sin parámetros ni valor de retorno:

```python
def fun():
    print("Welcome to GFG")

fun()
```

#### Ejemplo 2
Otra función de saludo:

```python
def saludo():
    print("Este es un saludo desde una función")

saludo()
```

### Ejercicios adicionales

#### Ejercicio 1.1: Función según hora
Escribe una función `saludar()` que imprima un saludo según la hora (valor interno de ejemplo):
- Buenos días si hora < 12
- Buenas tardes si 12 ≤ hora < 20
- Buenas noches si hora ≥ 20

```python
def saludar(hora):
    if hora <0 or hora > 24:
        print("Formato de hora incorrecto, debe ser un valor entre 0 y 24")
    elif hora < 12:
        print("Buenos días")
    elif 12 <= hora < 20:
        print("Buenas tardes")
    else:
        print("Buenas noches")

saludar(-1)

saludar(5)

saludar(16)

saludar(23)
```

```python
# Test ejercicio 1.1
import io, sys
cap = io.StringIO()
sys.stdout = cap
saludar()
sys.stdout = sys.__stdout__
assert cap.getvalue().strip() == 'Buenas tardes'
```

#### Ejercicio 1.2: Contar hasta 5
Define una función `contar_hasta()` que imprima los números del 1 al 5.

```python
def contar_hasta():
    for i in range(1, 6, 1):
        print(f"La i vale {i}")

contar_hasta()
```

```python
# Test ejercicio 1.2
import io, sys
cap = io.StringIO()
sys.stdout = cap
contar_hasta()
sys.stdout = sys.__stdout__
assert cap.getvalue().strip() == '1\n2\n3\n4\n5'
```

#### Ejercicio 1.3: Saludo personalizado
Define una función `saludo_personalizado(nombre)` que imprima 'Hola {nombre}!'

```python
def saludo_personalizado(nombre):
    print(f"Hola {nombre}!")

saludo_personalizado("Pablo")
```

```python
# Test ejercicio 1.3
import io, sys
cap = io.StringIO()
sys.stdout = cap
saludo_personalizado('Ana')
sys.stdout = sys.__stdout__
assert cap.getvalue().strip() == 'Hola Ana!'
```
