```python
# Ejercicio 1: Elementos básicos del programa
# Tema: Juego de energía para superhéroes
def calcular_energia_final(energia_base, factor):
    factor_minimo = 0.1
    if factor_minimo >= factor:
        energia_final = float(energia_base * factor_minimo)
    else:
        energia_final = float(energia_base * factor)
    return energia_final

#Tests
assert calcular_energia_final(100, 0.5) == 50.0
assert calcular_energia_final(100, 0.01) == 10.0 #Se aplica el mínimo
assert calcular_energia_final(0, 0.5) == 0
```

```python
# Ejercicio 2
# Tema: Detector de nível de peligro zombie
def nivel_peligro(cantidad):
    if cantidad == 0:
        return "Zona segura"
    elif cantidad >= 1 and cantidad <= 5:
        return "Peligro leve"
    elif cantidad >= 6 and cantidad <= 9:
        return "Peligro medio"
    #elif cantidad > 10
    #   return "¡Huye!" Si quisieramos añadir else como corrector de errores o algo
    else:
        return "¡Huye!"

#Tests
assert nivel_peligro(0) == "Zona segura"
assert nivel_peligro(3) == "Peligro leve"
assert nivel_peligro(9) == "Peligro medio"
assert nivel_peligro(25) == "¡Huye!"
```

```python
# Ejercicio 3: Mini juego
# Tema: Juego de adivinar el número secreto
def juego_adivinar(intentos=3, numero_secreto=7):
    while intentos > 0:
        intentouser = int(input("Introduzca el número secreto: "))
        if intentouser == numero_secreto:
            return "¡Correcto!"
            break
        elif intentouser != numero_secreto:
            intentos -= 1
    if intentos <= 0:
        return "Fallaste"


print(juego_adivinar(intentos=1))
print(juego_adivinar(intentos=2, numero_secreto=15))
print(juego_adivinar(intentos=1, numero_secreto=20))
```

```python
# Ejercicio 4: Control de excepciones
# Tema: Simulador de pociones mágicas
def leer_potencia(valor):
    try:
        return int(valor)
    except ValueError:
        return "Error: la potencia debe ser un número entero" 

#Tests
assert leer_potencia("25") == 25
assert leer_potencia("0") == 0
assert leer_potencia("abc") == "Error: la potencia debe ser un número entero"
```

```python
# Ejercicio 5: Funciones
# Tema: Mini-juego de carreras futuristas
def calcular_puntos(velocidad, choques):
    puntos = int((velocidad * 2) - (choques * 3))
    if puntos <= 0:
        return 0
    else:
        return puntos
#Tests
assert calcular_puntos(10, 2) == 14
assert calcular_puntos(5, 10) == 0
assert calcular_puntos(0, 0) == 0
print(calcular_puntos(100.75, 5.75)) #Con esta línea veo que se redondean y se convierten en una variable tipo int correctamente
```
