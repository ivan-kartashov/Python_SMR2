## 1. Adivina el número

```python
import random
numero_secreto = random.randint(1, 100)
while True:
    intento = int(input("Adivina el número (1-100): "))
    if intento < numero_secreto:
        print("Más alto.")
    elif intento > numero_secreto:
        print("Más bajo.")
    else:
        print("¡Correcto!")
        break
```

## 2. Piedra papel tijera

```python
import random
opciones = ["piedra", "papel", "tijera"]
cpu = random.choice(opciones)
user = input("Elige piedra, papel o tijera: ").lower()
print("CPU eligió:", cpu)
if user == cpu:
    print("Empate")
elif (user == "piedra" and cpu == "tijera") or (user == "papel" and cpu == "piedra") or (user == "tijera" and cpu == "papel"):
    print("Ganaste")
else:
    print("Perdiste")
```

## 3. Trivia

```python
puntos = 0
preguntas = {"¿Cuántos bits tiene un byte? ":"8","¿Qué protocolo usa la web? ":"http"}
for p, r in preguntas.items():
    if input(p).lower() == r:
        puntos += 1
print("Puntuación:", puntos, "de", len(preguntas))
```

## 4. Calculadora

```python
op = input("Introduce operación: + - * /: ")
a = float(input("Introduce número 1: "))
b = float(input("Introduce número 2: "))
if op == "+": print(a + b)
elif op == "-": print(a - b)
elif op == "*": print(a * b)
elif op == "/": print(a / b)
```

## 5. Carrera de dados

```python
import random
jugador = cpu = 0
while jugador < 20 and cpu < 20:
    input("Tirar dado...")
    jugador += random.randint(1,6)
    cpu += random.randint(1,6)
    print("Jugador:", jugador, "CPU:", cpu)
print("¡Ganaste!" if jugador>=20 else "Perdiste.")
```

## 6. Password Ninja

```python
pwd = input("Introduce contraseña: ")
cond1 = any(c.isupper() for c in pwd)
cond2 = any(c.islower() for c in pwd)
cond3 = any(c.isdigit() for c in pwd)
cond4 = len(pwd) >= 8
print("Contraseña válida" if cond1 and cond2 and cond3 and cond4 else "No válida")
```

## 7. Laberinto

```python
pos = 0
while pos != 3: #Posicion 3 en este caso sera la posicion que querremos que alance el usuario para batir el laberinto
    mov = input("Mover (izq/der): ").lower() #Aqui le pedimos al usuario que introduzca lo que necesitemos para moverse
    if mov == "der": pos+=1 #Al moverse a la derecha con der el usuario aumenta su posición en un 1
    elif mov == "izq" and pos > 0: pos-=1 #Al moverse a la izquierda no sucede nada, pero si la posicion es superior al uno esta disminuye
    print("Posición:", pos)
print("¡Salida encontrada!") #Se ejecuta cuando salimos del bucle y señala al usuario su victoria
```

```python
pos = 0
while pos != 3: #Posicion 3 en este caso sera la posicion que querremos que alance el usuario para batir el laberinto
    mov = input("Mover (< para izquierda/> para derecha): ").lower() #Aqui le pedimos al usuario que introduzca lo que necesitemos para moverse
    
    if mov == ">": pos+=1 #Si el usuario introduce la tecla de derecha, su posición aumenta en uno, este laberinto es más un camino
    elif mov == "<" and pos > 0: pos-=1 #Si el usuario usa la tecla de la izquierda retrocede un paso y su posicion disminuye en un 1
    print("Posición:", pos)
print("¡Salida encontrada!") #Se ejecuta cuando salimos del bucle y señala al usuario su victoria


#------------Avanzado-------------
#Para poder usar las teclas necesitariamos usar funciones que tendriamos que importar, la función keyboard sería útil.
#La siguiente manera de implementar algo así sería usando curses, como if key == curses.KEY_RIGHT elif curses.KEY_LEFT que también tendriamos que importar.
```

## 9. Memorama

```python
cartas = ["a", "b", "a", "b"]
mostrar = ["_"]*4
while "_" in mostrar:
    print(mostrar)
    i = int(input("Carta 1: "))
    j = int(input("Carta 2: "))
    if cartas[i] == cartas[j]:
        mostrar[i] = cartas[i]
        mostrar[j] = cartas[j]
        print("¡Match!")
    else:
        print("No coincide.")
print("¡Juego completado!")
```

## 10. Simulador de puntos

```python
puntos=0
acciones={"1":("Resolver ejercicio",10),"2":("Ayudar",20),"3":("Participar",5)}
while True:
    print("1) Ejercicio | 2) Ayuda | 3) Participación | 0) Salir")
    op=input("Elige: ")
    if op=="0": break
    accion,valor=acciones[op]
    puntos+=valor
    print(f"{accion} (+{valor}) Total:{puntos}")
print("Puntuación final:", puntos)
```
