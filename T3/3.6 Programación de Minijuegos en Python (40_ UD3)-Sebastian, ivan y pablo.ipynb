## 1. Adivina el número

```python
#Import hace funcion al random
import random
#Un numero al azar es elegido y tenemos que adivinarlo introduciendo numeros
numero_secreto = random.randint(1, 100)
intentos = 0
# While es un bucle de verdadero o falso que nos imprimira tres cosas
# "Mas alto" cuando el numero que ponemos es demasiado alto
# "Mas bajo" cuando el numero que ponemos es demasiado bajo
# "Correcto" cuando el numero es el correcto
while True:
    intentos +=1
    intento = int(input("Adivina el número (1-100): "))
    if intento < numero_secreto:
        print("Más alto.")
    elif intento > numero_secreto:
        print("Más bajo.")
    else:
        print("¡Correcto!")
        print ("Has necesitado " + str(intentos) + " Intentos")
        break
```

## 2. Piedra papel tijera

```python
import random
opciones = ["piedra", "papel", "tijera"]
vidas_jugador = vidas_cpu = 3
#Una lista de opciones que se usara despues
while vidas_jugador > 0 or vidas_cpu > 0:
    cpu = random.choice(opciones)
    # Eligue una opcion de las tres que tiene disponible en una lista de forma aleatoria
    # y tenemos que elegir una opcion de una lista
    user = input("Elige piedra, papel o tijera: ").lower()
    print("CPU eligió:", cpu)
    '''
    El If sirve para para que imprima un resultado:
     - Si user y cpu eligieron lo mismo, tienen "empate"
     - Si user elige una opcion que le gana a lo que eligio cpu entonces imprime "Ganaste"
     - Si user elige una opcion que le gana a lo que eligio cpu entonces imprime "Perdiste"
    '''
    if user == cpu:
        print("Empate")
    elif (user == "piedra" and cpu == "tijera") or (user == "papel" and cpu == "piedra") or (user == "tijera" and cpu == "papel"):
        print("Ganaste")
        vidas_cpu -= 1
        print ("Le quedan ", vidas_cpu)
    else:
        print("Perdiste")
        vidas_jugador -= 1
        print ("Te quedan ", vidas_jugador)
    if vidas_jugador < 0 or vidas_cpu < 0:
        continue
    elif vidas_jugador == 0 or vidas_cpu == 0:
        break
```

## 5. Carrera de dados

```python
import random
#Deja ambos "Jugadores" en el mismo inicio, el 0
jugador = cpu = 0
#Mientras ninguno de los jugadores tenga un valor mayor de veinte, se seguira ejecutando el while 
while jugador < 20 and cpu < 20:
    input("Tirar dado...")
    #Cada ves que se ejecuta, se tira un "dado" y lo que de cada uno se suma al valor de cada jugador
    jugador += random.randint(1,6)
    cpu += random.randint(1,6)
    print("Jugador:", jugador, "CPU:", cpu)
#Al dar 20 si el jugador tiene ese valor, gano, si no, perdio. No hay empate
print("¡Ganaste!" if jugador>=20 else "Perdiste.")
```

## 6. Password Ninja

```python
#Pide la introduccion de un valor que luego sera juzgado bajo ciertas reglas
pwd = input("Introduce contraseña: ")
#Cada condicion juzga a la contraseña
#isuper: Si hay mayusculas.
#islower: Minusculas.
#isdigit: Si hay digitos.
#len(pwd): cuenta la cantidad de caracteres que tiene la contraseña
cond1 = any(c.isupper() for c in pwd)
cond2 = any(c.islower() for c in pwd)
cond3 = any(c.isdigit() for c in pwd)
cond4 = len(pwd) >= 12
cond5 = any(len("?, !, _ #") for c in pwd)
#Si todas las condiciones se cumplen, entonces se imprime "Valida", si no, "No valida"
print (pwd)
print("Contraseña válida" if cond1 and cond2 and cond3 and cond4 and cond5 else "No válida")
```

## 7. Laberinto

```python
pos = 0
while pos != 3: #Posición 3 en este caso es la posición que el usuario tiene que alcanzar
    mov = input("Mover (izq/der): ").lower() #Aqui le pediremos al usuario el input que va a realizar y con el .lower le permitimos que no le importen las mayúsculas ni las minúsculas
    if mov == "der": pos+=1 #Cuando el usuario se mueve a la derecha con der, su posición aumenta
    elif mov == "izq" and pos > 0: pos-=1 #Mientras que al ir a la izquierda, si su posición no es nula este ira para atrás
    print("Posición:", pos)
print("¡Salida encontrada!") #Le decimos al usuario que ganó cuando salga del búcle
```

## 9. Memorama

```python
#Declaramos los valores en forma de lista
cartas = ["a", "b", "a", "b"]
mostrar = ["_"]*4
#Este bucle se encarga de emparejar los pares y dar una respuesta concluyente
while "_" in mostrar:
    print(mostrar)
    i = int(input("Carta 1: ")) # de 0 a 3
    j = int(input("Carta 2: ")) # de 0 a 3
    #meto primera posicion a y segunda posicion
    # aqui nos hace ver si tenemos match o no
    if cartas[i] == cartas[j]:
        mostrar[i] = cartas[i]
        mostrar[j] = cartas[j]
        print("¡Match!")
    else:
        print("No coincide.")
print("¡Juego completado!")
```

## 10. Simulador de puntos

```python
puntos=0 # Son los puntos con los que empezamos
acciones={"1":("Resolver ejercicio",10),"2":("Ayudar",20),"3":("Participar",5)}
while True:
    print("1) Ejercicio | 2) Ayuda | 3) Participación | 0) Salir")
    op=input("Elige: ")
    if op=="0": break # Si ponemos el 0 salimos
    accion,valor=acciones[op] # Se le asigna a cada accion una puntuacion
    puntos+=valor # Se le suma nuestros puntos al valor de la accion asignada
    print(f"{accion} (+{valor}) Total:{puntos}")
print("Puntuación final:", puntos)
```

# 11 Extra (para subir nota) Quiz aleatorio

```python
import random #Importamos la función random para que podamos elegir respuestas aleatoriamente

questions = {
    "¿En qué año fue fundado Madrid?": "865", "¿Cómo se llama el río que divide Alemania y Francia?": "Rin", "¿En qué año cayó Roma?": "476", "¿Quién escribió El Señor de los Anillos?": "Tolkien", "¿Cuántas vidas tiene un gato?": "7"
}
question_list = list(questions) #Convertimos el diccionario en una lista para que podamos usarla solo para la pregunta
question = random.choice(question_list)
correct_answer = questions[question] #Ahora, usamos el diccionario para averiguar que respuesta es la correcta
lives = 2   
while lives > 0:
    answer = input(question + " ")
    if answer.lower() == correct_answer.lower(): #En caso de que la respuesta del usuario coincida con la respuesta real
        print("Correcto")
        break
    else: #En el caso de que el usuario se equivoque
        lives -= 1
        if lives >= 0:
            print("Incorrecto Te quedan:", lives, "vidas")
if lives <= 0: #Si el usuario responde mal dos veces
    print("Has perdido")
else: #Si el usuario acierta
    print("Has Ganado!!!")
```
